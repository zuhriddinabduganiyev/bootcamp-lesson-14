```python
import pandas as pd
import numpy as np

group_cols = ['REGION', 'Ownership', "Degree_Type", "Locale_Short"]
def fillna_grouped(data, group_cols, column, method):
    if method == 'median':
        return data.groupby(by=group_cols)[column].transform(
            lambda x: x.fillna(np.nanmedian(x)) if not x.isnull().all() else x
        )
    elif method == 'mean':
        return data.groupby(by=group_cols)[column].transform(
            lambda x: x.fillna(np.nanmean(x)) if not x.isnull().all() else x
        )
    else:
        raise ValueError("Method must be either 'median' or 'mean'")
data['COMPL_RPY_5YR_RT_NUMERIC_FILLNA'] = fillna_grouped(data, group_cols, 'COMPL_RPY_5YR_RT_NUMERIC', 'median')
data
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
fig, ax = plt.subplots(figsize=(12, 8))
data['COMPL_RPY_5YR_RT_NUMERIC'].plot(kind='hist', ax=ax, alpha=0.5, color='blue', bins=20, label='Numeric')
data['COMPL_RPY_5YR_RT_NUMERIC_FILLNA'].plot(kind='hist', ax=ax, alpha=0.5, color='green', bins=20, label='Fillna_Median')
data['COMPL_RPY_5YR_RT_NUMERIC_FILLNA_MEAN'].plot(kind='hist', ax=ax, alpha=0.5, color='red', bins=20, label='Fillna_Mean')
ax.set_xlabel('Value')
ax.set_ylabel('Frequency')
ax.legend()
plt.grid()
plt.title('Histograms of NaN, Fillna_Median, Fillna_Mean')
plt.show()
```
